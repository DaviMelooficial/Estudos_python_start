```python
import os


#Dicionario com todos os carros do estoque
carros_estoque = {
     0 : "Chevrolet Tracker - R$ 120 / dia",
     1 : "Chevrolet Onix - R$ 90 / dia",
     2 : "Chevrolet Spin - R$ 150 / dia",
     3 : "Hyundai HB20 - R$ 85 / dia",
     4 : "Hyundai Tucson - R$ 120 / dia",
     5 : "Fiat Uno - R$ 60 / dia",
     6 : "Fiat Mobi - R$ 70 / dia",
     7 : "Fiat Pulse - R$ 130 / dia"
    }

#Dicionario vazio para receber os carros que forem alugados
carros_alugados = {
}

#Laço de repetição que garante a excecução do programa até que seja quebrado
while True:
    
    os.system("cls") #Função OS que limpa o terminal a cada execução do laço
    
    print("=============")
    print("Bem vindo à locadora de carros!")
    print("=============")
    print("O que deseja fazer?")
    print("")
    print("0 - Mostrar portifólio | 1 - Alugar um carro | 2 - Devolver um carro")
    
    acao = int(input()) #Variavel de ação no menu

    #Condicionais if e elif para cada possivel ação do menu
    if acao == 0:
        '''
        Utiliza de um for para printar cada item do dicionario 
        sendo representados por "i" e "carro"
        '''
        for i, carro in carros_estoque.items():
            print("[{}]".format(i), carro)
        print()
        print()
        print()
        print("=============")
        print("0 - CONTINUAR | 1 - SAIR")

        """
        Revisando este código percebo que a lógica de continuar e sair 
        após cada interação no menu poderia ter sido transformada em uma função
        levando em consideração o número de vezes que eu tive que escreve-la no código
        """
        acao_portfolio = int(input())
        if acao_portfolio == 0:
            pass
        elif acao_portfolio == 1:
            break
    elif acao == 1:
        for i, carro in carros_estoque.items():
            print("[{}]".format(i), carro)
        print()
        print()

        print("=============")
        print("Escolha o código do carro:")
        modelo = int(input()) #Recebe o modelo do carro para o cálculo
            
        print("Diga por quantos dias:") 
        dias = int(input()) #Recebe o número de dias para o cálculo

        '''
        REVISANDO ESTE CÓDIGO VEJO QUE
        ESTE ENORME IF PODERIA SER UMA ÚNICA FUNÇÃO
        '''
        if modelo == 0:
            result = dias * 120
        elif modelo == 1:
            result = dias * 90
        elif modelo == 2:
            result = dias * 150
        elif modelo == 3:
            result = dias * 85
        elif modelo == 4:
            result = dias * 120
        elif modelo == 5:
            result = dias * 60
        elif modelo == 6:
            result = dias * 70
        elif modelo == 7:
            result = dias * 130

        '''
        Print somente do nome do carro utilizando o método ".split()"
        para dividir em lista o item do dicionario referente ao modelo
        com indice 0 (o nome do carro que vem antes do " -")
        '''
        print("Você escolheu {} por {} dias.".format(carros_estoque[modelo].split(" -")[0], dias))
        print("O aluguel vai custar R$ {}. Deseja alugar?".format(result))
        print()
        print("0 - SIM | 1 - NÃO")
        acao_locacao = int(input())
        
        if acao_locacao == 0:
            print("Você alugou o {} por {} dias.".format(carros_estoque[modelo].split(" -")[0], dias))

            #Esta mecânica é a chave do código adiciono o item do diciniario de estoque ao dos carros alugados
            carros_alugados[modelo] = carros_estoque[modelo]
            #Removo o item do dicionario de estoque
            del carros_estoque[modelo]

            #Isto poderia ser uma função!
            print()
            print("=============")
            print("0 - CONTINUAR | 1 - SAIR")
            acao_portfolio = int(input())
            if acao_portfolio == 0:
                pass
            elif acao_portfolio == 1:
                break
        else:
            pass
    elif acao == 2:
        print("Segue a lista de carros alugados, qual você deseja devolver?")
        for i, carro in carros_alugados.items():
            print("[{}]".format(i), carro)
        print()
        print()
        print()
        print("Escolha o código do carro que deseja devolver:")
        modelo = int(input())

        #Mecânica inversa para manipulação dos dicionarios
        carros_estoque[modelo] = carros_alugados[modelo]
        del carros_alugados[modelo]
        
        print("Obrigado por devolver o carro {}".format(carros_estoque[modelo].split(" -")[0]))
        print("=============")
        print("0 - CONTINUAR | 1 - SAIR")
        acao_portfolio = int(input())
        if acao_portfolio == 0:
            pass
        elif acao_portfolio == 1:
            break
```
